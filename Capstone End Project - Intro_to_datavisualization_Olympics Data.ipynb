```python
# Total number of medals based on sex in the complete duration of the years?
# Solution 1-
sum_of_medals = df.groupby('Sex')['Medal'].sum()
Total_medals = pd.DataFrame({'Sex': sum_of_medals.index, 'Total Medals' : sum_of_medals.values})
print('Total number of Medals based on Sex:')
print(Total_medals)

# Solution 2-
sum_of_medals_1 = df.groupby('Sex')['Medal'].sum()
total_medals_1 = pd.DataFrame({'Sex': sum_of_medals.index, 'Total Medals' : sum_of_medals_1.values}).to_dict()
print('\nTotal number of Medals based on Sex:')
print("Total number of medals for Males:",total_medals_1['Total Medals'][1])
print("Total number of medals for Femals:",total_medals_1['Total Medals'][0])

# Total count of Males and Females in the complete duration of the years?
Count_of_gender = df.groupby('Sex')['Sex'].count()
total_count = pd.DataFrame({'Sex': Count_of_gender.index, 'Count of Gender' : Count_of_gender.values}).to_dict()
print('\nTotal count of Males and Females:')
print("Total number of Males:",total_count['Count of Gender'][1])
print("Total number of Femals:",total_count['Count of Gender'][0])
```

```python
# Create a new column as Sentiment in the dataframe and also should the % of each sentiment.
conditions = [(df['Medal'] >1), (df['Medal'] == 0), (df['Medal']==1)]
choices = ['Positive', 'Negative','Neutral'] #Positive is encoded to 1 and negative to 0
df['Sentiment'] = np.select(conditions,choices, default='Negative')
df['Sentiment'].value_counts(normalize=True)*100
```

```python
#Create a pie chart to showcase the sentiment distribution using matplotlib.
plt.figure(figsize = (6,6))
plt.pie(df['Sentiment'].value_counts(), labels = df['Sentiment'].value_counts().index, autopct='%1.1f%%')
plt.title('Pie Chart for Sentiment')
plt.show()
```

```python
# Perform a sentiment analysis such that athletes who won 1 medal were having neutral sentiment, 0 negative and more than 2 positive.
plt.figure(figsize=(7,7))
plt.pie(df["Sentiment"].value_counts(),labels=df["Sentiment"].value_counts().index,autopct='%1.1f%%', startangle = 90, shadow=True)
plt.legend(title ='Sentiment')
plt.show()
```

```python
# Further customize the plot by colour coding as below and also extract the positive piece by 30% away from the complete chart.
# Red - negative
# Green - positive
# Blue- Neutral
plt.figure(figsize=(7,7))
plt.pie(df["Sentiment"].value_counts(),labels=df["Sentiment"].value_counts().index,autopct='%1.1f%%', startangle = 90, shadow=True, colors = ['r', 'g', 'b'], explode =[0,0.3,0])
plt.legend(title ='Sentiment')
plt.show()
```
